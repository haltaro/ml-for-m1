# M1グランプリを因子分析

## import

```python
%matplotlib inline
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
sns.set(style="ticks")

# matplotlibで日本語を表示するための設定です．
from matplotlib.font_manager import FontProperties
font_path = '/usr/share/fonts/truetype/takao-gothic/TakaoPGothic.ttf'
font_prop = FontProperties(fname=font_path)
matplotlib.rcParams['font.family'] = font_prop.get_name()

# 因子分析（Factor analysis）
from sklearn.decomposition import FactorAnalysis
```

## Data

```python
df = pd.read_csv("data/2017.csv")
df
```

```python
sns.pairplot(df, hue="結果", palette="husl")
```

何となく，上沼さん，小朝さん，渡辺さんの評点が実際の順位と相関が強いように見えます．
しかしこのままでは，参加者同士にどのような相関関係があるのか，ひと目ではわかりにくいです．
そこで，**因子分析**を使い，審査結果を二次元空間に圧縮してみます．

## Factor analysis

```python
x = np.array(df.values)[:,1:-1]
t = np.array([1 if i=="通過" else 0 for i in df["結果"].values])
```

```python
x_fa = FactorAnalysis(n_components=2).fit_transform(x)
```

```python
x_fa
```

```python
for ii in range(len(t)):
    plt.plot(x_fa[ii, 0], x_fa[ii, 1], c= 'b' if t[ii] else 'r', marker='o',
             alpha=0.7)
    plt.annotate("{}\n{}点".format(df.iloc[ii, 0], sum(df.iloc[ii, 1:-1])),
                 x_fa[ii]+[.1, 0], )
```

## 振り返ってみる

過去のM1についても，同様の処理を行ってみます．`plot_fa_m1()`は，上記の一連の処理を一括して行う関数です．

```python
def plot_fa_m1(year=2017):
    df = pd.read_csv("data/{}.csv".format(year))
    x = np.array(df.values)[:,1:-1]
    t = np.array([1 if i=="通過" else 0 for i in df["結果"].values])
    x_fa = FactorAnalysis(n_components=2).fit_transform(x)
    
    for ii in range(len(t)):
        plt.plot(x_fa[ii, 0], x_fa[ii, 1], c= 'b' if t[ii] else 'r', marker='o',
                 alpha=0.7)
        plt.annotate("{}\n{}点".format(df.iloc[ii, 0], sum(df.iloc[ii, 1:-1])),
                     x_fa[ii]+[.1, 0], )
    
    return df
```
